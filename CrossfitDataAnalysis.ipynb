```python
import pandas as pd
import matplotlib
```

```python
rawData = pd.read_csv("df_open_2023.csv")
```

```python
rawData
```

```python
trimmedData = rawData.drop(columns=["countryOfOriginName", "countryOfOriginCode", "affiliateId", "affiliateName", "status", "year", "gender"])
```

```python
trimmedData["heightType"] = trimmedData["height"].str[-2:]
trimmedData["weightType"] = trimmedData["weight"].str[-2:]

trimmedData = trimmedData[(trimmedData["height"].notna()) & (trimmedData["weight"].notna())]
trimmedData = trimmedData.drop_duplicates(subset=["overallRank"])

trimmedData["height"] = trimmedData["height"].str.extract("(\d+)")
trimmedData["weight"] = trimmedData["weight"].str.extract("(\d+)")

trimmedData["height"] = trimmedData.apply(lambda row: int(row.height) if row.heightType == 'in' else round(int(row.height) / 2.52), axis=1)
trimmedData["weight"] = trimmedData.apply(lambda row: int(row.weight) if row.weightType == 'lb' else round(int(row.weight) * 2.2), axis=1)
```

```python
trimmedData = trimmedData[trimmedData["height"] >= 36]
trimmedData["bmi"] = trimmedData.apply(lambda row: round((row.weight * 703) / (row.height ** 2), 2), axis=1)
```

```python
maleTrimmedData = trimmedData[trimmedData["genderId"] == 1]
```

```python
maleTrimmedData = maleTrimmedData.drop(columns=["heightType", "weightType", "competitorName", "firstName", "lastName"])
```

```python
hrMaleTrimmedData = maleTrimmedData[maleTrimmedData["overallRank"] <= 50000]
```

```python
hrMaleTrimmedData["bmi"] = hrMaleTrimmedData.apply(lambda row: round((row.weight * 703) / (row.height ** 2), 2), axis=1)
```
